```python
import numpy as np
import matplotlib.pyplot as plt
with open("settings.txt", "r") as settings:
    tmp = [float(i) for i in settings.read().split("\n")]
data_array = np.loadtxt("data.txt", dtype = int)
fig, ax = plt.subplots(figsize = (16, 10), dpi = 250)
voltage_array = data_array * tmp[0]
x = np.arange(len(data_array))
t = x * tmp[1]
ax.plot(t, voltage_array, label = 'V(t)', c = 'r', linestyle = '-', linewidth = 1, marker = 'o', markersize = 2, markevery = 15)
ax.set_xlabel(", ")
ax.set_ylabel(", ")
ax.set_title("    RC ", ha="center", va="center", wrap=True )
ax.set_xlim([0, t.max()+0.2])
ax.set_ylim([0, voltage_array.max()+0.2])
plt.minorticks_on()
plt.grid(which='major')
plt.grid(which='minor', linestyle='--')
i = 0
istr = str(i*tmp[1])
istr1 = str((x.max()-i)*tmp[1])
f = ['  = ', str(len(data_array)*0.01), '']
f_str = " ".join(f)
ax.text(6, 2.2, f_str, fontsize=8)
ax.legend()
fig.savefig("test.svg")
plt.show()
```
